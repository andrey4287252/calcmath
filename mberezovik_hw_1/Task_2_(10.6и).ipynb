```python
import numpy as np
from numpy import linalg as LA
import math
```

```python
#Размер матриц NxN
N = 100
A = np.zeros((N,N))
F = np.arange(1,101,1.0)
for i in range(N):
    for j in range(min(i+2,N)):
        if (i != j):
            A[i][j] = (i+1)/(j+1)
        else:
            A[i][j] = 100
F
```

# Якоби

```python
def yakob(A,X,F,N):
    Y = np.zeros(N)
    for i in range(N):
        for j in range(N):
            if j == i:
                continue
            Y[i] += -(X[j]*(A[i][j]))
        Y[i] += F[i]
        Y[i] = Y[i]/(A[i][i])
    X = Y
    return X
```

```python
X = np.zeros(N)
for i in range(10):
    X = yakob(A,X,F,N)
    print(X)
```

```python
print(F - A@X)
```

# Зейдель

```python
def zeid(A,X,F,N):
    for i in range(N):
        X[i] = 0
        for j in range(N):
            if j == i:
                continue
            X[i] += -(X[j]*(A[i][j]))
        X[i] += F[i]
        X[i] = X[i]/(A[i][i])
    return X
```

```python
L = np.tril(A, -1)
U = np.triu(A, 1)
D = A-U-L
R = -U@LA.inv(L+D)
q = LA.norm(R,np.inf)
```

```python
# eps - ошибка
eps = 0.00000001
Y = np.zeros(N)
Y = zeid(A,Y,F,N)
while True:
    Z = Y.copy()
    Y = zeid(A,Y,F,N)
    d = LA.norm(Z-Y,np.inf)
    if(d < eps*(1-q)):
        break
print(Y)
```

```python
print(F - A@Y)
```

```python
#Скалярное произведение двух векторов
def scal(x,y):
    m = x@(y.transpose())
    return m
```

```python
u = F
for j in range(10000):
    u = A@u/LA.norm(u, 2)
print("lambda",scal(u,A@u)/scal(u,u))
```

```python
#Кубическая:
print('Кубическая:',LA.norm(A, np.inf)* LA.norm(LA.inv(A),np.inf))
#Октаэдрическая:
print('Октаэдрическая:',LA.norm(A, 1)* LA.norm(LA.inv(A), 1))
```

```python
B = A@A.transpose()
u = F
for j in range(20):
    u = B@u/LA.norm(u, 2)
lamba = math.sqrt(scal(u,B@u)/scal(u,u))
```

```python
C = LA.inv(A) @ LA.inv(A).transpose()
u = F
for j in range(100):
    u = C@u/LA.norm(u, 2)
lamba_inv = math.sqrt((scal(u,C@u)/scal(u,u)))
print("Евклидова:",lamba_inv * lamba)
```

# Метод Гаусса

```python
def gauss_1(A,F,N):
    C = A.copy()
    L = F.copy()
    for i in range(N):
        for j in range(i+1,N):
            coef = C[j][i]/C[i][i]
            L[j] -= L[i]*coef
            for k in range(i,N):
                C[j][k] -= C[i][k]*coef
    return C, L
```

```python
B,L = gauss_1(A,F,N)
```

```python
def gauss_2(B,L,N):
    D = B.copy()
    K = L.copy()
    for i in reversed(range(N)):
        for j in range(0,i):
            coef = D[j][i]/D[i][i]
            K[j] -= K[i]*coef
            for k in range(0,i+1):
                D[j][k] -= D[i][k]*coef
    return D, K
```

```python
R, J = gauss_2(B,L,N)
Sol = np.zeros(N)
for i in range(N):
    Sol[i] = J[i]/R[i][i]
```
