```python
import numpy as np
from numpy import linalg as LA
import math
```

```python
#Размер матриц NxN
N = 100
A = np.zeros((N,N))
F = np.array([i+1 for i in range(N)])
for i in range(N):
    for j in range(min(i+2,N)):
        if (i != j):
            A[i][j] = (i+1)/(j+1)
        else:
            A[i][j] = 100
A
```

# Зейдель

```python
def yakob(A,X,F,N):
    Y = np.zeros(N)
    for i in range(N):
        for j in range(N):
            if j == i:
                continue
            Y[i] += -(X[j]*(A[i][j]))
        Y[i] += F[i]
        Y[i] = Y[i]/(A[i][i])
    X = Y
    return X
```

```python
X = np.zeros(N)
for i in range(10):
    X = yakob(A,X,F,N)
    print(X)
```

```python
print(F - A@X)
```

$\textbf{Метод Зейделя}$

```python
def zeid(A,X,F,N):
    for i in range(N):
        X[i] = 0
        for j in range(N):
            if j == i:
                continue
            X[i] += -(X[j]*(A[i][j]))
        X[i] += F[i]
        X[i] = X[i]/(A[i][i])
    return X
```

```python
X = np.zeros(N)
for i in range(10):
    X = zeid(A,X,F,N)
    print(X)
```

```python
print(F - A@X)
```

```python
#Скалярное произведение двух векторов
def scal(x,y):
    m = x@(y.transpose())
    return m
```

```python
u = F
for j in range(1000000):
    u = A@u/LA.norm(u, 2)
print("lambda",scal(u,A@u)/scal(u,u))
```

```python
#Кубическая:
print('Кубическая:',LA.norm(A, np.inf)* LA.norm(LA.inv(A),np.inf))
#Октаэдрическая:
print('Октаэдрическая:',LA.norm(A, 1)* LA.norm(LA.inv(A), 1))
```

```python
B = A@A.transpose()
u = F
for j in range(20):
    u = B@u/LA.norm(u, 2)
lamba = math.sqrt(scal(u,B@u)/scal(u,u))
```

```python
C = LA.inv(A) @ LA.inv(A).transpose()
u = F
for j in range(100):
    u = C@u/LA.norm(u, 2)
lamba_inv = math.sqrt((scal(u,C@u)/scal(u,u)))
print("Евклидова:",lamba_inv * lamba)
```
