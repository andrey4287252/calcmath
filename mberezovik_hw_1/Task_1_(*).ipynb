# Первый способ вычисления

```python
def f(x,n):
    x = (2**n) * math.sqrt(2*(1-math.sqrt(1-(x/(2**(n)))**2)))
    return x
```

```python
p = 2.
for i in range(1,61):
    p = f(p,i)
    print("[",i,"]",p)
```

# Второй способ вычисления

```python
def g(x):
    x = x/(2 + math.sqrt(4 - x))
    return x
```

```python
r = 2/(2+math.sqrt(2))
p = 4 * math.sqrt(r)
print("[",3,"]",p)
for i in range(4,61):
    r = g(r)
    p = (2**(i-1)) * math.sqrt(r)
    print("[",i,"]",p)
```
