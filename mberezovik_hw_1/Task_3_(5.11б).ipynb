```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def f(x):
        return 1/x
```

```python
x = np.arange(1,2,0.01)
y = np.array([f(l) for l in x])
N = np.size(x)
```

```python
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = N
sum_x2 = np.sum([l**2 for l in x])
```

```python
a = (N*sum_xy - sum_x*sum_y)/(N*sum_x2 - (sum_x)**2)
b = (sum_y - a*sum_x)/N
```

```python
fig = plt.figure()
plt.plot(x,y)
plt.plot(x, a*x + b)
```
