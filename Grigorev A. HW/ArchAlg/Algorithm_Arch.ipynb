```python
def rec(p, n):
    return (2**n)*((2*(1 - (1 - (p/2**n)**2)**0.5))**0.5)

def cycle(n):
    p = 2.
    for i in range(1, n + 1, 1):
        p = rec(p, i)
    return p
```

```python
for i in range(1, 61, 1):
    print("i: ", i)
    print("p: ", cycle(i))
```
