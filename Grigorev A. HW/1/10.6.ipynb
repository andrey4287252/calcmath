```python
import numpy as np
import pandas as pd
import functools
import math
import fractions
import itertools as it
from numpy import linalg as LA
```

```python
def zeid(A,X,F,N):
    for i in range(N):
        X[i] = 0
        for j in range(N):
            if j == i:
                continue
            X[i] += -(X[j]*(A[i][j]))
        X[i] += F[i]
        X[i] = X[i]/(A[i][i])
    return X
```

```python
def yakob(A,X,F,N):
    Y = np.zeros(N)
    for i in range(N):
        for j in range(N):
            if j == i:
                continue
            Y[i] += -(X[j]*(A[i][j]))
        Y[i] += F[i]
        Y[i] = Y[i]/(A[i][i])
    X = Y
    return X
```

```python
#Скалярное произведение двух векторов
def scal(x,y):
    m = x@(y.transpose())
    return m
```

```python
F = np.array([i for i in range(1, 101, 1)])
A = 10*np.identity(100) + np.diag([1 for i in range(99)], k = -1) + np.diag([1 for i in range(99)], k = 1) + 0.1*np.diag([1 for i in range(98)], k = 2)
Y = np.array([0. for i in range(1, 101, 1)])
Z = np.array([0. for h in range(1, 101, 1)])
for j in range(10):
    Y = yakob(A, Y, F, 100)
for k in range(10):
    Z = zeid(A, Z, F, 100)
print(A@Y - F)
print(A@Z - F)
#print(np.linalg.solve(A, F))
```

$\textbf{Максимальное и минимальное собственные числа:}$

```python
B = A@np.transpose(A)
u = np.array([1 for i in range(100)])
for j in range(10000):
    print("Шаг №:", j+1)
    print("lambda",scal(u,B@u)/scal(u,u))
    u = B@u
    u = u/LA.norm(u, 2)
LA.eig(B)
```

```python
С = np.linalg.inv(A)@np.transpose(np.linalg.inv(A))
u = np.array([1 for i in range(100)])
for j in range(10000):
    print("Шаг №:", j+1)
    print("lambda",scal(u,С@u)/scal(u,u))
    print("1/lambda", scal(u,u)/scal(u,С@u))
    u = С@u
    u = u/LA.norm(u, 2)
```
